```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('degreedays.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['country'].value_counts()
```

```python
sns.set_theme()
sns.set(rc={'figure.figsize':(12, 10)})
sns.scatterplot(data=df, x='date', y='hdd_15.5', hue='country')
```

```python
sns.set_theme(style='whitegrid')
sns.set(rc={'figure.figsize':(12, 10)})
sns.scatterplot(data=df, x='date', y='hdd_15.5', hue='latitude', size=1, palette=sns.color_palette("crest", as_cmap=True))
```

```python
sns.lineplot(data=df, x='date', y='hdd_15.5', hue='country')
```

```python
from datetime import datetime
df['datetime'] = df['date'].apply(lambda x: datetime.fromisoformat(x))
```

```python
sns.kdeplot(data=df, x='datetime', y='hdd_15.5')
```
